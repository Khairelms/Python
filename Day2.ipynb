```python
a = 19
b = 34
c = 23
d = 141
e = 67

largest = 0

if a>largest:     #   19 > 0 (t)
  largest = a     #   largest  = 19

if b>largest:     #   34 > 19 (t)
  largest = b     #   largest  = 34

if c>largest:     #   23 > 34 (f)
  largest = c     #   largest  = 34

if d>largest:     #   141 > 34 (t)
  largest = d     #   largest  = 141

if e>largest:     #   67 > 141 (f)
  largest = e     #   largest  = 141

print(f'Largest is {largest}')
```

```python
# Day 4
# While Loop

# while loop is used to repeat a block of code until condition remains TRUE
# while first will check the condition , if the condition is TRUE
# while loop will execute the body
# while will check the condition again
# if the condition is still TRUE while again executes the body

# syntax
# initialization
# while condition:
#   code to be executed
#   incr / decr

i = 1    # initialization            #  i --> 1      again condn will be checked        again condn will be checked
while i<=3:                          #  1<=3 (t)           2<=3 (t)                          3<=3(t)                         4<=3(F)X
  print(i)                           #  print(1)           print(2)                          print(3)
  i = i + 1                          #  i = 2              i = 3                             i = 4
```

```python
# Palindrome Of A Number

num = 123                       # num = 123      num = 12       num = 1
save = 0                        # save = 0       save = 3       save = 32

rem = num % 10                  # rem =   3       rem = 2       rem = 1
save = save * 10 + rem          # save =  3       save = 32     save = 321
num = num // 10                 # num = 12        num = 1       num = 0
```

```python
num = int(input('Insert Number :'))
copy_num = num
save = 0

while num!=0:
  rem = num % 10
  save = save * 10 + rem
  num = num // 10

print(f'Reverse Of A Number : {save}')

if save == copy_num:
    print('Palindrome Number')
else:
    print('Not Palindrome Number')
```

```python
# break and continue statement

# break statement
# break it is used to terminate the execution of a Loop [for/while] in between
# break should be defined inside decision making

for i in range(1,11):
  if i == 6:
    break
  print(i)
```

```python
# continue statement
# continue it is used to skip the current execution of a Loop [for/while]

for i in range(1,11):
  if i == 6 or i == 9:
    continue
  print(i)
```
