```python
# OOPS
# OOP stands for Object-Oriented Programming.
# Object-oriented programming (OOP) is a programming paradigm based on the
# concept of "objects"
 
# Procedural programming is about writing procedures or methods that perform operations
# on the data, while object-oriented programming is about creating objects
# that contain both data and methods.
 
# Object-oriented programming has several advantages over procedural programming:
 
# a.) OOP is faster and easier to execute
# b.) OOP provides a clear structure for the programs
# c.) OOP helps to keep the Java/Python code DRY "Don't Repeat Yourself", and makes the code easier
#     to maintain, modify and debug
# d.) OOP makes it possible to create full reusable applications with less code and
#     shorter development time
 
# Class
# A class is a blueprint that defines the variables and the methods common to all objects.
# a class is a collection of methods and variables
 
 
# Object
# An object is called an instance of a class.
 
 
# Self Keyword
# The self parameter is a reference to the current instance of the class, and is used to
# access variables that belongs to the class.
# The first parameter of each method in a class must be the self ,
# which refers to the calling object. However, you can give any name to the first
# parameter, not necessarily self.
 
# __init__ in Python
#  Constructors are used to initializing the objectâ€™s state. The task of constructors is
# to initialize(assign values) to the data members of the class when
# an object of the class is created.
# The __init__() function is called automatically every time the class is being used to create
# a new object.
```

```python
# class
# a class is a collection of variables and methods
# variables defined inside class is known as Class Variables / Member Variables
# methods/functions defined inside class is known as Class Methods / Member Methods
# in python we define class using class keyword

class myclass:
    '''
    This is my first class in Python OOP
    contoh tujuan tentang class ini
    '''

    a = 10            # member variable
    b = 20

    def func1():        # member method
        print('I am the method of a class')

    def func2():
        print('I am the another method of a class')
```

```python
# accessing class variables and methods

class myclass:

    a = 10            
    b = 20

    def func1():       
        print('I am the method of a class')

    def func2():
        print('I am the another method of a class')

# Rule 1
# in order to access class variables and methods
# use classname with dot operator
# dot operator --> member access operato

print(myclass.a)
print(myclass.b)
                               # not used at all, used Rule 2
myclass.func1()
myclass.func2()
```

```python
# Rule 2
# in order to access class variables and methods define an object for a Class
# a class can have n number of objects
 
class myclass:
 
    a = 10 
    b = 20
 
    # self represent the current calling object
    # whenever you are defining a function inside a class , first parameter needs to be SELF
    def func1(self):                                        
        print(' I am the method of a class')                # func1(obj1) giving your class a clear idea that now obj1 wants to use this function
                                                            # func1(obj2) giving your class a clear idea that now obj2 wants to use this function
    def func2(self):
        print(' I am the another method of a class')
 
    def func3(self,val1,val2):                         # func3(obj1,10,20)      func3(obj2,30,40)
        print(f'{val1} and {val2}')
 
obj1 = myclass()  # syntax for creating an object of a class
obj2 = myclass()  # creating another object
 
print(obj1.a)
print(obj1.b)
 
print(obj2.a)
print(obj2.b)
 
obj1.func1()
obj2.func1()
 
obj1.func3(10,20)
obj2.func3(30,40)
```

```python
# Task
# Take 3 information from 3 users
# user1  name   mobno   email
# user2 name   mobno   email
# user3 name   mobno   email
 
# print all information
# print the information of only second user   : user2  name   mobno   email
# print the name and mobno of 3rd user

userInfo = []

for i in range(1,4):       #  1  2  3              # i --> 1
    print(f'\nEnter Details of User {i}')
    name = input('Enter Your Name: ')              # name --> abd
    mn = input('Enter Your Mobile No: ')           # mn --> 123
    email = input('Enter Your Email: ')            # email --> abc@g.com

    userInfo.append([name,mn,email])               # [[name, mn,email],[name, mn,email],[name, mn,email]]
```

```python
print("ALL Information")
for i in userInfo:
    print(f'{i[0]}, {i[1]}, {i[2]}')
   
print("\nSecond User Information")
print(userInfo[1][0], userInfo[1][1], userInfo[1][2])

print("\nName and Mobile no of 3rd User")
print(userInfo[2][0], userInfo[2][1])
```

```python
# WITH USING __init__
 
class userInfo:
 
    def __init__(self,name,mobno,email):    # u1,"John",123,"j@gmail.com"        u2,"Alex",456,"al@gmail.com"   u3,"Smith",789,"sm@gmail.com
        self.n = name                       # u1.n = "John"                      u2.n = "Alex"                  u3.n = "Smith"
        self.mn = mobno                     # u1.mn = 123                        u2.mn = 456                    u3.mn = 789
        self.e = email                      # u1.e = "j@gmail.com"               u2.e = "al@gmail.com"          u3.e = "sm@gmail.com"    
 
    def printUserInfo(self):                 # printUserInfo(u1)                 printUserInfo(u2)           
        print(self.n,self.mn,self.e)         #  u1.n  u1.mn   u1.e                 u2.n  u2.mn  u2.e
 
 
u1 = userInfo("John",123,"j@gmail.com")
u2 = userInfo("Alex",456,"al@gmail.com")
u3 = userInfo("Smith",789,"sm@gmail.com")
 
print("All Information")
u1.printUserInfo()
u2.printUserInfo()
u3.printUserInfo()
 
print("\nPrint the information of only second user ")
u2.printUserInfo()
 
print("\nPrint the name and mobno of 3rd user ")
print(u3.n , u3.mn)
```

```python
# Procedural + OOPs

class userInfo:

    def __init__(self,name,mobno,email):   
        self.n = name                      
        self.mn = mobno                    
        self.e = email                       

    def printUserInfo(self):                       
        print(self.n,self.mn,self.e)       

objects = []  # Empty List

for i in range(3):
    print(f"\nEnter Details Of User {i+1}")
    uname = input("Enter Your Name : ")
    umn = input("Enter Your Mobile No. : ")
    uem = input("Enter Your Email : ")

    obj = userInfo(uname,umn,uem)
    objects.append(obj)

# Print All Information 


# Print the information of only second user


# Print the name and mobno of 3rd user
```
